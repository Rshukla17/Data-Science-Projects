```python
df.dropna(inplace = True,ignore_index = True)
```

```python
df['genre'] = df['genre'].apply(lambda x:x.replace('Sci-Fi','scifi'))
```

```python
df['genre'].sample(5)
```

```python
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
```

```python
import nltk
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
corpus = []
for i in range(len(df)):
    l = re.sub('[^a-zA-Z]',' ',df['description'][i])
    l = l.split()
    l = [wnl.lemmatize(i) for i in l if i not in stopwords.words('english')]
    l = ' '.join(l)
    corpus.append(l)
   
```

```python
df['description'] = corpus
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf.fit_transform(df['description']).toarray()
```

```python
featured_vectors = pd.DataFrame(tfidf.fit_transform(df['description']).toarray(),columns = tfidf.get_feature_names_out())
featured_vectors.columns
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(featured_vectors, featured_vectors)
similarity
```

```python
list_of_all_titles = df['title'].tolist()
list_of_all_titles
```

# Enter the name of the movie or show


```python
movie_name = 'The Big Bang Theory'
```

```python
import difflib

find_close_match = difflib.get_close_matches(movie_name,list_of_all_titles)

find_close_match
```

```python
close_match = find_close_match[0]
close_match
```

```python
index_of_movie = df[df.title == close_match].index.values[0]
```

```python
similarity_score = list(enumerate(similarity[index_of_movie]))
similarity_score
```

```python
sorted_score = sorted(similarity_score,key = lambda x : x[1],reverse = True)[1:11]
sorted_score
```

```python
k = 1
for i in sorted_score:
    m = df.iloc[i[0]]['title'].capitalize()
    print(f'{k}.{m}')
    k += 1
```
